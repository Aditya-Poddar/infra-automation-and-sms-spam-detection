```python
import re

def analyze_chat(chat_text):
    regex_pattern = r'(\d{2}/\d{2}/\d{2}), (\d{2}:\d{2}\s*[ap]m) - (.+): (.+)'
    messages = re.findall(regex_pattern, chat_text)
    
    chat_analysis = {
        'total_messages': len(messages),
        'participants': [],
        'message_count': {},
        'most_active_participant': '',
        'links': []
    }
    
    for message in messages:
        date = message[0]
        time = message[1]
        sender = message[2]
        content = message[3]
        
        # Track participants
        if sender not in chat_analysis['participants']:
            chat_analysis['participants'].append(sender)
        
        # Count messages per participant
        if sender not in chat_analysis['message_count']:
            chat_analysis['message_count'][sender] = 1
        else:
            chat_analysis['message_count'][sender] += 1
        
        # Track most active participant
        if chat_analysis['most_active_participant'] == '' or \
           chat_analysis['message_count'][sender] > chat_analysis['message_count'][chat_analysis['most_active_participant']]:
            chat_analysis['most_active_participant'] = sender
        
        # Extract links
        urls = re.findall(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', content)
        chat_analysis['links'].extend(urls)
    
    return chat_analysis

# Example usage
chat_text = """
29/08/22, 11:42 am - Kulvir Sir: Hello everyone! Please join the Zoom meeting: https://us06web.zoom.us/j/86263031696?pwd=elI2WkJ2L2I0SkJYQVFXUVh3b3gwZz09
29/08/22, 11:45 am - John Doe: Hi Kulvir! I will join the meeting.
29/08/22, 11:47 am - Mary Smith: Me too! Looking forward to it.
"""

analysis_result = analyze_chat(chat_text)

print("Total Messages:", analysis_result['total_messages'])
print("Participants:", analysis_result['participants'])
print("Message Count:", analysis_result['message_count'])
print("Most Active Participant:", analysis_result['most_active_participant'])
print("Links:", analysis_result['links'])
```
